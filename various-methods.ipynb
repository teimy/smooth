### Importing
First of all we need to import all python stuff we use as data monkeys:

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Genering data
Let us generate some data! And noise:

```python
x = np.linspace(0,np.pi,500)
y = np.cos(x)
noise = np.random.randn(500)/10

plt.figure(figsize=(20,10))
plt.plot(x,y+noise)
plt.plot(x,y)
plt.show()
```

Pseudo-absorption line:

```python
def abs_line(x, mean, sigma, scale):
    return np.exp(-scale*np.exp(-((x-mean)*sigma)**2))
```

Now same with absoption line:

```python
x = np.linspace(0,np.pi,500)
y = np.cos(x)+1
noise = np.random.randn(500)/10

plt.figure(figsize=(20,10))
plt.plot(x,y*abs_line(x,1,100,0.2)+noise)
plt.plot(x,y)
plt.show()

plt.figure(figsize=(20,10))
plt.plot(x,y*abs_line(x,1,100,0.2)+noise)
plt.plot(x,y*abs_line(x,1,100,0.2))
plt.show()
```

Some non-vectoristed spaghetti code:

```python
x = np.linspace(0,np.pi,500)
y = np.cos(x)+1
noise = np.random.randn(500)/10

with_lines = np.copy(y)

for mean, sigma, scale in zip(np.abs(np.random.rand(20))*3, 50+np.abs(np.random.rand(20))*100, np.random.rand(20)*0.7):
    with_lines *= abs_line(x, mean, sigma, scale)

plt.figure(figsize=(20,10))
plt.plot(x,with_lines+noise)
plt.plot(x,with_lines)
plt.plot(x,y)
plt.show()

```
