Now let us smooth:

```python
x = np.linspace(0,np.pi,500)
y = np.cos(x)
noise = np.random.randn(500)/10

kernel = np.ones(20)

y_smoothed = np.convolve(y+noise, kernel, mode='valid')/np.sum(kernel)
x_smoothed = np.convolve(x, kernel, mode='valid')/np.sum(kernel)


plt.figure(figsize=(20,10))
plt.plot(x,y+noise)
plt.plot(x,y)
plt.plot(x_smoothed, y_smoothed)
plt.xlabel('Pseudo-wavelength', fontsize=24)
plt.ylabel('Pseudo-intensity', fontsize=24)
plt.legend(['Data','Initial sinus', 'Smoothed by sliding average with window size=20'], fontsize=24)
plt.savefig('pics/sliding-average.png')

plt.show()
```

Pseudo-absorption line:

```python
def abs_line(x, mean, sigma, scale):
    return np.exp(-scale*np.exp(-((x-mean)*sigma)**2))
```

Now same with absoption line:

```python
x = np.linspace(0,np.pi,500)
y = np.cos(x)+1
noise = np.random.randn(500)/10

with_lines = np.copy(y)

for mean, sigma, scale in zip(np.abs(np.random.rand(20))*3, 50+np.abs(np.random.rand(20))*100, np.random.rand(20)*0.7):
    with_lines *= abs_line(x, mean, sigma, scale)


    
plt.figure(figsize=(20,10))
plt.plot(x,with_lines+noise)
plt.plot(x,with_lines)
for degree in [2, 3, 4, 5]:
    y_savgol = sig.savgol_filter(with_lines, 201, degree)
    plt.plot(x,y_savgol)
    
plt.legend(['Data with noise', 'Data without noise', 'Savgol 2 degree', 'Savgol 3 degree', 'Savgol 4 degree', 'Savgol 5 degree'], fontsize=24)
plt.savefig('pics/savgol-lines.png')
plt.show()

```

Preparing function:

```python
import rpy2.robjects as robjects
import rpy2
from rpy2.robjects.packages import STAP
from rpy2.robjects import numpy2ri
numpy2ri.activate()

with open('afs.r', 'r') as f:
    string = f.read()
    
myfunc = STAP(string, "AFS")

def envelope(wavelength, data, q=0.95, d=0.25, normalize=False):
    """ Function for making envelope for spectra from echelle grating
    
    Default q and d taken from source; @teimy considers q~0.5 and d~0.7 to be
    best values fitting almost everything nice
    Generally, the bigger signal-to-noise leads to bigger amounts of q
    More details at https://github.com/xinxuyale/AFS
    """
    if not(normalize):
        return data/np.array(myfunc.AFS(np.array([wavelength, data]).T, q=q, d=d))
    else:
        tmp = (data - data.mean())/(data.max()-data.min()) # different normalization
        tmp += 1.5*np.abs(np.min((data - data.mean())/data.std()))
        env = tmp/np.array(myfunc.AFS(np.array([wavelength, tmp]).T, q=q, d=d))
        env -= 1.5*np.abs(np.min((data - data.mean())/data.std()))
        return env*(data.max()-data.min()) + data.mean()
```

```python
x = np.linspace(0,np.pi,500)
y = np.cos(x)+1
noise = np.random.randn(500)/10

with_lines = np.copy(y)

for mean, sigma, scale in zip(np.abs(np.random.rand(20))*3, 50+np.abs(np.random.rand(20))*100, np.random.rand(20)*0.7):
    with_lines *= abs_line(x, mean, sigma, scale)

env = envelope(x, with_lines, q=0.7, d=0.5, normalize=True)    
    
plt.figure(figsize=(20,10))
plt.plot(x,with_lines+noise)
plt.plot(x,with_lines)
plt.plot(x,env)
    

plt.savefig('pics/afs.png')
plt.show()

```
